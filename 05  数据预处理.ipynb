# 四、处理分类变量

## 1. n个分类变量转换为<n个变量

```python
data_
```

```python
thick_ = np.arange(1,11)
thick_
```

```python
thick_category = ['thin','thin','thin',
                'medium','medium','medium','medium',
                'thick','thick','thick',]
```

```python
thick_category_ = ['thin']*3 +['medium']*4 + ['thick']*3
thick_category_
```

```python
dict2 = dict(zip(thick_,thick_category_))
dict2
```

```python
data_['Thickness_class'] = data_['Clump Thickness'].map(dict2)
```

```python
data_
```

## 2. 连续变量转换为分类变量

```python
bins = [0,4,8,11]
```

```python
labels = ['small','medium','big']
```

```python
pd.cut(data_['Uniformity of Cell Shape'],bins,labels=labels)
```

```python
data_['Cell Size'] = pd.cut(data_['Uniformity of Cell Shape'],bins,labels=labels)
```

```python
data_
```

## 3. 分类变量转换为数字或哑变量

```python
data_['Thickness_classnum'] = data_['Thickness_class'].replace(to_replace=['thin','medium','thick'],value=[1,2,3])
```

```python
data_
```

```python
pd.get_dummies(data_['Cell Size'])
```

```python
data_=pd.concat((data_,pd.get_dummies(data_['Cell Size'],prefix = 'Cell Size ')),axis=1)
```

```python
data_
```

```python
data_.set_index('ID',inplace=True)
data_
```

```python
new_index = np.random.permutation(data_.index)
```

```python
data_ = data_.reindex(index = new_index)
data_
```

```python
data_.columns
```

```python
data_ = data_.reindex(columns=['Clump Thickness', 'Uniformity of Cell Size',
       'Uniformity of Cell Shape', 'Marginal Adhesion',
       'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin',
       'Normal Nucleoli', 'Mitoses', 'Class',
       'Thickness_classnum', 'Cell Size _small', 'Cell Size _medium',
       'Cell Size _big', 'Thickness_class', 'Cell Size'])
data_
```
