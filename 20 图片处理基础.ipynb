# 一、Matplotlib图片处理
        （导入/裁切/缩放/灰度/保存）

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
```

```python
img = image.imread("input/R_LAO.jpg")
```

```python
plt.imshow(img)
```

```python
img
```

```python
img.shape
```

```python
(1296-734)/2
```

```python
281+734
```

```python
img_new = img[:,281:1015]
```

```python
img_new.shape
```

```python
plt.imshow(img_new)
```

```python
img_small = img_new[::2,::2]
```

```python
img_small.shape
```

```python
plt.imshow(img_small)
```

```python
gravity= np.array([0.299,0.587,0.114])
img_gray = np.dot(img_small, gravity)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
image.imsave('output/R_LAO_mtp.jpg',img_gray, cmap="gray")
```

# 二、PIL图片处理

```python
import PIL
from PIL import Image
```

```python
im = Image.open("input/R_LAO.jpg")
```

```python
im
```

```python
im_new = im.crop((281,0,1015,734))
im_new
```

```python
im_small = im_new.resize((367,367))
im_small
```

```python
im_gray = im_small.convert(mode = "L")
im_gray
```

```python
im_gray = np.array(im_gray)
```

```python
im_gray
```

```python
im_gray.shape
```

```python
im_gray = Image.fromarray(im_gray)
im_gray
```

```python
im_gray.save("output/R_LAO_PIL.jpg")
```
