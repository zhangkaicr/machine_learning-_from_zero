# 三、自编码器

```python
physical_devices = tf.config.list_physical_devices('GPU')
```

```python
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
conv_encoder = keras.models.Sequential([
    keras.layers.InputLayer(input_shape=[256, 256,3]),
    keras.layers.Conv2D(4, kernel_size=3, padding="same",activation="selu"),
    keras.layers.Conv2D(8, kernel_size=3, padding="same",activation="selu"),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Conv2D(16, kernel_size=3, padding="same",activation="selu"),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Conv2D(16, kernel_size=3, padding="same",activation="selu"),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Flatten()])
```

```python
conv_encoder.summary()
```

```python
conv_decoder = keras.models.Sequential([
    keras.layers.Reshape([32, 32, 16],input_shape=[16384]),
    keras.layers.Conv2DTranspose(16, kernel_size=3, strides=2, padding="same",activation="selu"),
    keras.layers.Conv2DTranspose(16, kernel_size=3, strides=2, padding="same",activation="selu"),
    keras.layers.Conv2DTranspose(8, kernel_size=3, strides=2, padding="same",activation="selu"),
    keras.layers.Conv2DTranspose(4, kernel_size=3, padding="same",activation="selu"),
    keras.layers.Conv2DTranspose(3, kernel_size=3, padding="same",activation="selu")])
```

```python
conv_decoder.summary()
```

```python
conv_ae = keras.models.Sequential([conv_encoder, conv_decoder])
```

```python
conv_ae.compile(loss='mse',metrics='mae',
                optimizer=keras.optimizers.RMSprop(learning_rate=0.0001))
```

```python
conv_ae.fit(train_x_,train_x_,epochs=200,batch_size=1,
         validation_data=(valid_x_,valid_x_),
         validation_batch_size=1)
```

# 四、建立模型

```python
model = keras.Sequential()
model.add(keras.layers.InputLayer( input_shape = train_codings.shape[1:]))
model.add(keras.layers.Dense(5000,activation='selu'))
model.add(keras.layers.Dense(2500,activation='selu'))
model.add(keras.layers.Dense(1000,activation='selu'))
model.add(keras.layers.Dense(500,activation='selu'))
model.add(keras.layers.Dense(100,activation='selu'))
model.add(keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', metrics = 'accuracy', 
             optimizer=keras.optimizers.SGD(learning_rate=0.0001))
```

```python
model.fit(train_codings,train_stenosis,batch_size=10,epochs=500,
          validation_data=(valid_codings,valid_stenosis),validation_batch_size=10)
```
