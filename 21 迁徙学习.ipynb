# 一、数据导入

```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
```

```python
list_1 = os.listdir('input/552/RCA/')
list_1
```

```python
img_552= []
for img_ in list_1:
    imgname = 'input/552/RCA/' + img_
    img = image.load_img(imgname,target_size=(512,512))
    img = image.img_to_array(img)
    img_552.append(img)
```

```python
train_x_ = train_x.astype("float32") * 1/255
train_x_ = train_x_[:,::2,::2]
print(train_x_.dtype)
print(train_x_.shape)
```

```python
valid_x_ = valid_x.astype("float32") * 1/255
valid_x_ = valid_x_[:,::2,::2]
print(valid_x_.dtype)
print(valid_x_.shape)
```

# 三、建立模型

```python
from tensorflow.keras import layers, models
```

```python
from tensorflow.keras.applications import InceptionV3
conv_base = InceptionV3(include_top=False, weights="imagenet", input_shape=(256,256,3))
model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation ='selu'))
model.add(layers.Dense(1, activation ='sigmoid'))
```

```python
conv_base.summary()
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', metrics='acc')
```

```python
conv_base.trainable = False
model.fit(train_x_,train_stenosis,epochs=50,batch_size=10,
         validation_data=(valid_x_,valid_stenosis),
         validation_batch_size=10)
```

```python
keras.utils.plot_model(conv_base)
```

```python
conv_base.trainable = True
top = ['activation_85','activation_91','activation_92','activation_93','concatenate_1','mixed10']
for layer in conv_base.layers:
    if layer.name in top:
        layer.trainable = True
    else:
        layer.trainable = False  
```

```python
tensorboard_cb = keras.callbacks.TensorBoard(log_dir='output/logs_',histogram_freq=1)
```

```python
model.compile(loss='binary_crossentropy', metrics='acc')
model.fit(train_x_, train_stenosis, 
    epochs=50, batch_size=10,
    validation_data=(valid_x_, valid_stenosis), 
    validation_batch_size=10,
    callbacks = [tensorboard_cb])
```

# 四、数据增强

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    horizontal_flip=True,
    vertical_flip=True,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2)
```

```python
model.fit(train_datagen.flow(train_x_, train_stenosis, batch_size=10),epochs=50, 
    validation_data=(valid_x_, valid_stenosis), 
    validation_batch_size=10)
```
